***Bivariate analysis*** - represent relatioship between two variables


**Types**

  1. line plot
  2. scatter plot
  3. heatmap
  4. regplot
  5. jointplot
  6. Bar
  7. kde
  8.box
  9.violin
  10.strip
  11.swarm
  12. cross tab
  13. stacked bar plot
  14. count plot
  15. pairplot


```python

# line plot - 
plt.figure(figsize = (18,8))
sns.lineplot(x = df['day'],y = df['tip'],palette='rainbow',hue = df['sex'])
plt.show()
```

```python
# scatter plot

plt.figure(figsize = (18,8))
sns.scatterplot(x=df['total_bill'],y =df['tip'],hue = df['sex'],palette = 'rainbow')
plt.title('scatter plot for total -> tip')
plt.show()
```

```python
plt.figure(figsize = (18,8))
sns.scatterplot(x = df['size'],y = df['tip'],hue = df['sex'],palette = 'rainbow')
plt.title('scatter plot for total -> size')
plt.show()
```

```python
plt.figure(figsize =(15,10))
sns.heatmap(df.corr(),annot = True,cmap='rainbow')
plt.show()
```

```python
# joint plot - represent scatter and histogram of two variables

plt.figure(figsize =(18,10))
sns.jointplot(x=df['total_bill'],y=df['tip'],hue = df['sex'],palette = 'crest')
plt.show()
```

```python
plt.figure(figsize= (15,10))
sns.barplot(x = df['sex'],y = df['total_bill'],hue = df['time'],palette = 'rainbow')
plt.show()
```

```python
# kde -A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, 
#analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

plt.figure(figsize=(15,10))
sns.kdeplot(x = df['total_bill'],hue = df['sex'],fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,)
plt.show()
```

```python
# boxplot

plt.figure(figsize=(15,10))
sns.boxplot(y =df['total_bill'],x = df['sex'],hue = df['time'],palette = 'crest')
plt.show
```

```python
# violin plot


plt.figure(figsize=(15,10))
sns.violinplot(y =df['total_bill'],x = df['sex'],hue = df['time'],palette = 'crest')
plt.show
```

```python
# strip

plt.figure(figsize=(15,10))
sns.stripplot(y =df['total_bill'],x = df['sex'],hue = df['time'],palette = 'rainbow')
plt.show
```

```python
# swarm

plt.figure(figsize=(15,10))
sns.swarmplot(y =df['total_bill'],x = df['sex'],hue = df['time'],palette = 'rainbow')
plt.show
```

```python
# boxenplot



plt.figure(figsize=(15,10))
sns.boxenplot(y =df['total_bill'],x = df['sex'],hue = df['time'],palette = 'crest')
plt.show
```

```python
#point plot

plt.figure(figsize=(15,10))
sns.pointplot(y =df['total_bill'],x = df['sex'],hue = df['sex'],palette = 'crest')
plt.show
```

```python

plt.figure(figsize=(15,10))
pd.crosstab(df['sex'],df['time']).plot(kind = 'bar')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(15,10))
sns.pairplot(df,palette = 'crest')
plt.show()
```
