***Bivariate analysis*** - represent relatioship between two variables


**Types**

1. Numerical - Numerical

  1. line plot
  2. scatter plot
  3. heatmap
  4. regplot
  5. jointplot

2. Numerical - Categorical

  1. Bar
  2. kde
  3.box
  4.violin
  5.strip
  6.swarm

3. Categorical - Categorical

 1. cross tab
 2. stacked bar plot
 3. count plot
 4. pairplot


```python

# line plot - 

sns.lineplot(x = df['day'],y = df['tip'],palette='rainbow')
```

```python
# scatter plot

sns.scatterplot(x=df['total_bill'],y =df['tip'],palette = 'rainbow')
plt.title('scatter plot for total -> tip')
plt.show()
```

```python
sns.scatterplot(x = df['size'],y = df['tip'],palette = 'rainbow')
plt.title('scatter plot for total -> size')
plt.show()
```

```python
plt.figure(figsize =(15,10))
sns.heatmap(df.corr(),annot = True,cmap='rainbow')
plt.show()
```

```python
# reg plot

plt.figure(figsize =(15,10))
sns.regplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

```python
# joint plot - represent scatter and histogram of two variables

plt.figure(figsize =(15,10))
sns.jointplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

```python
# Numerical - Categorical
df.head()
```

```python
plt.figure(figsize= (15,10))
sns.barplot(x = df['sex'],y = df['total_bill'],palette = 'rainbow')
plt.show()
```

```python
# kde -A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, 
#analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

plt.figure(figsize=(15,10))
sns.kdeplot(x = df['total_bill'],hue = df['sex'],fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,)
plt.show()
```

```python
# boxplot

plt.figure(figsize=(15,10))
sns.boxplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
# violin plot


plt.figure(figsize=(15,10))
sns.violinplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
# strip

plt.figure(figsize=(15,10))
sns.stripplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
# swarm

plt.figure(figsize=(15,10))
sns.swarmplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
# boxenplot



plt.figure(figsize=(15,10))
sns.boxenplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
#point plot

plt.figure(figsize=(15,10))
sns.pointplot(y =df['total_bill'],x = df['sex'],palette = 'crest')
plt.show
```

```python
# cross tab

plt.figure(figsize=(15,10))
pd.crosstab(df['sex'],df['time']).plot(kind = 'bar')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(15,10))
sns.pairplot(df,palette = 'crest')
plt.show()
```
