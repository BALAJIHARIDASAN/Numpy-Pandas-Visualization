```python
df_num.columns
```

#***Histogram***- 

```python
def hist_data():
  '''function for histogram of numerical dataset'''
  for x in df_num.columns:
    sns.histplot(df[x])
    plt.title(f'histogram for {x}')
    plt.show()
    print('_______________')
    print('skewness :',df[x].skew())
    if df[x].skew()>0:
      print('right skew')
    elif df[x].skew()<0:
      print('left skew')
    else:
      print('no skew')
```

```python
hist_data()
```

#***BOXPLOT***

```python
def Box_data():
  '''function for histogram of numerical dataset'''
  for x in df_num.columns:
    sns.boxplot(df[x])
    plt.title(f'Boxplot for {x}')
    plt.show()
    print('_______________')
    print('skewness :',df[x].skew())
    if df[x].skew()>0:
      print('right skew')
    elif df[x].skew()<0:
      print('left skew')
    else:
      print('no skew')

    print('------------------------------------------')
```

```python
Box_data()
```

```python
def Dist_data():
  '''function for displot of numerical dataset'''
  for x in df_num.columns:
    sns.displot(df[x])
    plt.title(f'distplot for {x}')
    plt.show()
    print('_______________')
    print('skewness :',df[x].skew())
    if df[x].skew()>0:
      print('right skew')
    elif df[x].skew()<0:
      print('left skew')
    else:
      print('no skew')

    print('------------------------------------------')
```

```python
Dist_data()
```
